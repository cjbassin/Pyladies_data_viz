```python
import make_graphs_all as mg
import pandas as pd
import numpy as np
from altair import *
```

```python
filename = 'Seattle_Pet_Licenses.csv'
animals = mg.load_clean_data(filename)
anim = mg.setup_bar_names(animals)
```

```python

alt_data = mg.write_to_altair_data(anim,["Name",'License Number'])
```

```python
Chart(Data(values=alt_data),
        description='Bar Chart',
    ).mark_bar().encode(
        x='Name:O',
        y='License Number:Q',
    )
```

```python
aniscat = mg.setup_scatter(animals)
```

```python

alt_data = mg.write_to_altair_data(aniscat,["Cat",'Dog','Livestock'])
```

```python
Chart(Data(values=alt_data), 
     description = 'Scatter').mark_point().encode(
    x='Cat:Q',
    y='Dog:Q',
    color='Livestock:N',
)
```
